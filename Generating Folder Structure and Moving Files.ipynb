```python
# Extracting ZIP from tool placed in Tool Bucket and placing F&D files in directory QC Test and SKU Trackers QC
from zipfile import ZipFile
import glob
import re
import shutil


flag=None
data,fail = [],[]
# path = input("Enter Tool Bucket path:  ")
# path.replace('\\','\\\\')
path = "C:\\Users\\Kumar.Anurag\\Downloads\\tool bucket"

titles = glob.glob(os.path.join(path, "*.zip"))
print(str(len(titles))+" files found in Tool Bucket")
print(*titles, sep="\n")

# destination = input("Enter Zip Extraction Destination QC Test Path (Month Name): ")
# destination.replace('\\','\\\\')
destination = "C:\\Users\\Kumar.Anurag\\OneDrive - GlobalData PLC\\Desktop\\F&D backup"

# temp = input("Enter SKU Trackers QC Path: ")
# temp.replace('\\','\\\\')
temp = "C:\\Users\\Kumar.Anurag\\OneDrive - GlobalData PLC\\Desktop\\Tracker - QC"

# Monthly_QC_filtered_path = temp+"\\Filtered"
# Monthly_QC_discarded_path = temp+"\\Discarded"
titles = glob.glob(os.path.join(path, "*.zip"))

for k,v in clients.items():
    for t in titles:
        for t1 in v:
            if re.search(t1,t):
                copy_path = destination+"\\"+k+"\\"+t1
                print("\nExtracting "+t1+"...",end = "\t")
                try:
                    with ZipFile(t) as z:
                        z.extractall(copy_path)
                        print("✓")
                        data.append(t)
                        filtered_path = copy_path+"\\Filtered"
                        discarded_path = copy_path+"\\Discarded"
                        filtered_files = glob.glob(os.path.join(filtered_path, "*.csv"))
                        discarded_files = glob.glob(os.path.join(discarded_path, "*.csv"))
                        print("\tCopying F&D Files ("+k+")...", end="\t")
                        Monthly_QC_filtered_path = temp+"\\"+k+"\\2023\\April\\Files Work Up\\Filtered"
                        Monthly_QC_discarded_path = temp+"\\"+k+"\\2023\\April\\Files Work Up\\Discarded"
                        for f in filtered_files:
                            shutil.copy2(f, Monthly_QC_filtered_path)
                        print(str(len(filtered_files))+" Filtered ✓", end=" ")
                        for f in discarded_files:
                            shutil.copy2(f, Monthly_QC_discarded_path)
                        print(str(len(discarded_files))+" Discarded ✓")
                            
                except Exception as e:
                    print("Invalid file: "+t+" ❌")
                    print(e)
                    fail.append(t)
            
```
